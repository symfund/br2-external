```python
import tensorflow as tf
from tensorflow.python.client import device_lib

print("=== General Hardware Checks ===")
print(f"TensorFlow Built with CUDA: {tf.test.is_built_with_cuda()}")
print(f"TensorFlow Built with GPU: {tf.test.is_built_with_gpu_support()}")

print("\n=== Available Compute Devices ===")
devices = tf.config.list_physical_devices()
for dev in devices:
    print(f" - {dev}")

print("\n=== Detailed Device Specifications ===")
local_device_protos = device_lib.list_local_devices()
for x in local_device_protos:
    if x.device_type == "GPU":
        print(f"Device Name: {x.name}")
        print(f"Memory Limit: {x.memory_limit / (1024**3):.2f} GB")
        print(f"Description: {x.physical_device_desc}\n")
```
