```python
import tensorflow as tf
import time

# Prevent TensorFlow from hogging all VRAM immediately if running parallel tasks
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

size = 20000
print(f"Generating random matrix arrays of size: {size} x {size}...")

# Force execution on CPU
print("\n--- Running on CPU ---")
with tf.device('/CPU:0'):
    cpu_matrix1 = tf.random.normal([size, size])
    cpu_matrix2 = tf.random.normal([size, size])
    
    start_time = time.time()
    cpu_output = tf.matmul(cpu_matrix1, cpu_matrix2)
    cpu_time = time.time() - start_time
    print(f"CPU Execution Time: {cpu_time:.4f} seconds")

# Force execution on GPU
print("\n--- Running on GPU ---")
with tf.device('/GPU:0'):
    # Warm-up pass to initialize CUDA context cleanly
    warm1 = tf.random.normal([100, 100])
    warm2 = tf.random.normal([100, 100])
    _ = tf.matmul(warm1, warm2)
    
    gpu_matrix1 = tf.random.normal([size, size])
    gpu_matrix2 = tf.random.normal([size, size])
    
    start_time = time.time()
    gpu_output = tf.matmul(gpu_matrix1, gpu_matrix2)
    gpu_time = time.time() - start_time
    print(f"GPU Execution Time: {gpu_time:.4f} seconds")

print("\n=== Benchmark Results ===")
print(f"Your GPU is {cpu_time / gpu_time:.2f}x faster than your CPU for large matrix workflows!")
```
